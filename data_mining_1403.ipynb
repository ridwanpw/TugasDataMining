```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = pd.read_csv("dataset.csv")
dataset.shape
```

```python
dataset.head()
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
dataset
```

```python
# reshape data as per group
data = dataset.groupby(['ID FAKTUR', 'BARANG'])['QTY'].sum().unstack().reset_index().fillna(0).set_index('ID FAKTUR')
data
```

```python
# change value of each element to 0 and 1
data = data.apply(lambda x:[1 if y >=1 else 0 for y in x])
data
```

```python
frequent_itemsets = apriori(data, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
rule = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rule
```
